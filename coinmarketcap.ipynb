```python
# Import libraries
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('http://www.coinmarketcap.com')
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
table = soup.find("table")
for tr in table.find_all('tr')[1:]:
    td = tr.find_all("td")
    p = td[2].find_all('p')
    try:
        text = p[1].text
        print(f"{text}")
    except IndexError:
        print(td[2].find_all("span")[1].text)
    #for td in td:
     #   print(td.prettify())
```
